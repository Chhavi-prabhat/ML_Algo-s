LINEAR REGRESSION implementation steps

1.Data gathering                                     
2.Data pre-processing                                 
3.Splitting the dataset into training and testing dataset.                                               
4.Training the model                                  
5.Predicting results based on user input                 
6.Visualizing the results predicted                   
7.Validation of linear regression model

```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#loading dataset and its pre-processing
df=pd.read_csv('student-scores.csv', usecols=['absence_days','weekly_self_study_hours','math_score'])
df

# dividing dataset
x=df[['absence_days','weekly_self_study_hours']]
y=df['math_score']

# spliting dataset
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2)

# importing LinearRegression
mod= LinearRegression()

#training model
mod.fit(X_train, y_train)

#prediction
y_pred=mod.predict(X_test)

#user_input
absent=int(input("Enter number of absent days: "))
study= float(input("Enter number of study hours: "))
user_ip=[[absent, study]]
prediction= mod.predict(user_ip)
print(f"Your predicted marks will be: {prediction[0]}")
```

```python
#to check accuracy by plotting study hours vs marks
plt.scatter(X_test['weekly_self_study_hours'], y_test,color='blue', label='Actual marks')
plt.scatter(X_test['weekly_self_study_hours'], y_pred, color='red', label='Predicted Marks')
plt.xlabel('Weekly study hours')
plt.ylabel('Marks')
plt.title('Linear Regression: Study hours vs marks')
plt.legend()
plt.show()
```

```python
#to check accuracy by plotting absent days vs marks
plt.scatter(X_test['absence_days'], y_test,color='blue', label='Actual marks')
plt.scatter(X_test['absence_days'], y_pred, color='red', label='Predicted Marks')
plt.xlabel('Absent days')
plt.ylabel('Marks')
plt.title('Linear Regression: Study hours vs marks')
plt.legend()
plt.show()
```

IMPLEMENTATION STEPS

5.   Predicting

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
iris=sns.load_dataset('iris', usecols=['petal_length','species'])
iris = iris[iris['species'].isin(['setosa', 'virginica'])]
iris['species'].replace({'setosa': 0, 'virginica': 1}, inplace=True)
# iris
x=iris[['petal_length']]
y=iris['species']
X_train,X_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
model=LogisticRegression()
model.fit(X_train, y_train)
pred=model.predict(X_test)
petal_len=float(input("Enter the length of petal: "))
user_input=[[petal_len]]
user_pred=model.predict(user_input)
print(user_pred[0])
if user_pred[0]==0:
  print("The predicted species is Setosa.")
else:
  print("The predicted species is Virginica.")
```

```python
# Visualization
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color='red',label='actual data')
plt.scatter(X_test, pred, color='blue', label='predicted data')
plt.show()
```

```python
# Validation of model
# y=1/1+e^-x
#1/1+2.7^-4.6
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.2f}")
```

**LOGISTIC REGRESSION FOR MULTICLASS CLASSIFICATION**

IMPLEMENTATION STEPS

1.Data gathering

2.Data processing

3.Dataset spliting

4.training model

5.Predicting

6.Visualizing the predictions and actual result

7.Validation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
iris=sns.load_dataset('iris')
# iris
iris['species'].unique()
iris['species'].replace({'setosa':0, 'virginica': 1, 'versicolor':2},inplace=True)
# iris
x=iris[['sepal_length','sepal_width', 'petal_length','petal_width']]
y=iris['species']
X_train,X_test,y_train, y_test=train_test_split(x,y,test_size=0.2)
model=LogisticRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
petal_w=float(input("Enter petal width: "))
petal_l=float(input("Enter petal length: "))
sepal_w=float(input("Enter sepal width: "))
sepal_l=float(input("Enter sepal length: "))
usr_input=[[petal_w,petal_l, sepal_w, sepal_l]]
prediction=model.predict(usr_input)
print(prediction[0])
if prediction[0]==0:
  print("The predicted species is Setosa")
elif prediction[0]==1:
  print("The predicted species is Virginica")
else:
  print("The predicted species is Versicolor")
```

```python
# Visualization
import matplotlib.pyplot as plt
plt.scatter(X_train['petal_length'], y_train, color='blue', label='Actual value')
plt.scatter(X_test['petal_length'], pred, color='red', label='Predicted value')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train['petal_width'], y_train, color='blue', label='Actual value')
plt.scatter(X_test['petal_width'], pred, color='red', label='Predicted value')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train['sepal_width'], y_train, color='blue', label='Actual values')
plt.scatter(X_test['sepal_width'], pred, color='red', label='Predicted values')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train['sepal_length'], y_train, color='blue', label='Actual flower')
plt.scatter(X_test['sepal_length'], pred, color='red', label='Predicted flower')
plt.legend()
plt.show()
```

```python
# Validation
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.2f}")
```
