LINEAR REGRESSION implementation steps

1.Data gathering                                     
2.Data pre-processing                                 
3.Splitting the dataset into training and testing dataset.                                               
4.Training the model                                  
5.Predicting results based on user input                 
6.Visualizing the results predicted                   
7.Validation of linear regression model

```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#loading dataset and its pre-processing
df=pd.read_csv('student-scores.csv', usecols=['absence_days','weekly_self_study_hours','math_score'])
df

# dividing dataset
x=df[['absence_days','weekly_self_study_hours']]
y=df['math_score']

# spliting dataset
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2)

# importing LinearRegression
mod= LinearRegression()

#training model
mod.fit(X_train, y_train)

#prediction
y_pred=mod.predict(X_test)

#user_input
absent=int(input("Enter number of absent days: "))
study= float(input("Enter number of study hours: "))
user_ip=[[absent, study]]
prediction= mod.predict(user_ip)
print(f"Your predicted marks will be: {prediction[0]}")
```

```python
#to check accuracy by plotting study hours vs marks
plt.scatter(X_test['weekly_self_study_hours'], y_test,color='blue', label='Actual marks')
plt.scatter(X_test['weekly_self_study_hours'], y_pred, color='red', label='Predicted Marks')
plt.xlabel('Weekly study hours')
plt.ylabel('Marks')
plt.title('Linear Regression: Study hours vs marks')
plt.legend()
plt.show()
```

```python
#to check accuracy by plotting absent days vs marks
plt.scatter(X_test['absence_days'], y_test,color='blue', label='Actual marks')
plt.scatter(X_test['absence_days'], y_pred, color='red', label='Predicted Marks')
plt.xlabel('Absent days')
plt.ylabel('Marks')
plt.title('Linear Regression: Study hours vs marks')
plt.legend()
plt.show()
```
